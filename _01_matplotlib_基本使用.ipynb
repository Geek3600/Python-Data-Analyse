```python
from matplotlib import pyplot as plt
```

```python
# 简单绘图


# 设置点
x = range(2, 26, 2)
y = [15,13,14.5,17,20,25,26,26,24,22,18,15]

#   

# 绘制图片
plt.plot(x,y,'c')

# 保存图像
plt.savefig("22.jpg")


# 展示图片
plt.show()
```

```python
# 设置图片大小

fig = plt.figure(figsize=(40,20), dpi=80)
# figsize：图像大小
# dpi (dot per int): 每英寸的像素点的个数

x = range(2,26,2)
y =  [15,13,14.5,17,20,25,26,26,24,22,18,15]
plt.plot(x,y)


plt.show()
```
